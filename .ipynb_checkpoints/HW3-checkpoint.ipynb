```python
import pandas as pd
import numpy as np

file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
purchase_data.df = purchase_data

purchase_data.head()
```

```python
totalplayernumber = purchase_data.df["SN"].nunique()
print("Total Number of Players:", totalplayernumber)
```

```python
avg_price = purchase_data.df["Price"].mean()
purchase_id = purchase_data.df["Purchase ID"].nunique()
unique_itemnames = purchase_data.df["Item Name"].nunique()
avg_age = purchase_data.df["Age"].mean()
item_id = purchase_data.df["Item ID"].nunique()
total_gender = purchase_data.df["Gender"].count()
male = purchase_data.df["Gender"].value_counts()['Male']
female = purchase_data.df["Gender"].value_counts()['Female']
binary = total_gender - male - female

summary_table = pd.DataFrame({"Purchase ID": [purchase_id],
                                   "Total Count": [totalplayernumber],
                                   "Avg. Age": [avg_age],
                                   "Male %": [male/total_gender],
                              "Female %": [female/total_gender],
                                   "Unique Item ID": [item_id],
                                   "Unique Item Name": [unique_itemnames],
                                   "Avg.Price": [avg_price]})
summary_table = summary_table.round(2)
summary_table
```

```python

gender_table = pd.DataFrame({"Gender": ["Male", "Female", "Binary"], "Total Count": [male, female, binary], "Percentage of Players":[female/total_gender, male/total_gender, binary/total_gender]})

gender_table = gender_table.set_index('Gender')
gender_table
```

```python
avp_price = purchase_data.df["Price"].count()
male_avp_price = avp_price/male
female_avp_price = avp_price/female
binary_avp_price = avp_price/binary
gender_summary = pd.DataFrame({"Gender": ["Female", "Male", "Binary"], "Purchase Count": [female, male, binary], "Average Purchase Price": [female_avp_price, male_avp_price, binary_avp_price], "Total Purchase Value": [female, male, binary],"Avg Total Purchase Value Per person": [female, male, binary] })
gender_summary = gender_summary.set_index("Gender")
gender_summary

```
